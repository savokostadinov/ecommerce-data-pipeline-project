```python
# skrejpa so izminuvanje na sekoja strana
import re
from urllib.parse import urlparse
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

def build_markets(driver, wait):
    driver.get("https://ramstore.com.mk/marketi/")
    stores = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.store")))
    markets = {}
    for store in stores:
        onclick = store.find_element(By.CSS_SELECTOR, "button[onclick]") \
                       .get_attribute("onclick")
        m = re.search(r"location\.href='([^']+)'", onclick)
        if not m:
            continue
        href = m.group(1)
        seg  = urlparse(href).path.rstrip("/").split("/")[-1]     
        slug = seg.replace("ramstore-", "").replace("-", "_")
        markets[slug] = href
    return markets

def scrape_market(driver, wait, slug, url):
    print(f"→ Scraping {slug} at {url}")
    driver.get(url)
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table.dataTable")))

    rows_data = []
    while True:
        rows = driver.find_elements(By.CSS_SELECTOR, "table.dataTable tbody tr")
        for row in rows:
            cells = row.find_elements(By.TAG_NAME, "td")
            rows_data.append({
                "ИМЕ НА МАРКЕТ":      slug,
                "НАЗИВ НА ПРОИЗВОД":        cells[0].text.strip(),
                "ПРОДАЖНА ЦЕНА":  cells[1].text.strip(),
                "ЕДИНЕЧНА ЦЕНА":  cells[2].text.strip(),
                "ОПИС НА ПРОИЗВОД": cells[3].text.strip(),
                "ДОСТАПНОСТ НА ПРОИЗВОД":   cells[4].text.strip(),
                "РЕДОВНА ЦЕНА":     cells[5].text.strip(),
                "ЦЕНА СО ПОПУСТ":  cells[6].text.strip(),
                "ПОЕНИ":      cells[7].text.strip(),
                "ТИП НА АКЦИЈА":  cells[8].text.strip(),
                "ВРЕМЕТРАЊЕ НА АКЦИЈА":  cells[9].text.strip(),
            })

        # Odenje na sledna strana do kraj i dokolku nema sledna brejknuva
        try:
            nxt = driver.find_element(By.CSS_SELECTOR, "a.paginate_button.next")
            if "disabled" in nxt.get_attribute("class"):
                break
            nxt.click()
            wait.until(EC.staleness_of(rows[0]))
        except (NoSuchElementException, IndexError):
            break

    return rows_data

options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {"profile.managed_default_content_settings.images": 2})
driver = webdriver.Chrome(options=options)
wait   = WebDriverWait(driver, 10)


markets = build_markets(driver, wait)
print("Достапни маркети", markets.keys())

# zemanje na vardar marketot
url_ramstore_vardar = markets["vardar"]

# skrejpanje preku funkcijata scrape_market
data_vardar = scrape_market(driver, wait, "vardar", url_ramstore_vardar)

driver.quit()

# zacuvuvanje vo csv format preku pandas bibliotekata
df = pd.DataFrame(data_vardar)
df.to_csv("ramstore_vardar.csv", index=False, encoding="utf-8")
print(f" Scraped {len(df)} rows for 'vardar' and saved to ramstore_vardar.csv")
```
