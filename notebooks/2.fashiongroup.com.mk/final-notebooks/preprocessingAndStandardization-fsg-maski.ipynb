```python
import pandas as pd
```

```python
raw_data = pd.read_csv('fashiongroup_maski.csv')
raw_data
```

```python
raw_data.isnull().sum()
```

```python
raw_data['Discount Price'] = raw_data['Discount Price'].fillna('0 MKД')
```

```python
raw_data.isnull().sum()
```

```python
raw_data = raw_data.dropna(subset=['Brand','Code','Description'])
```

```python
raw_data.head(5)
```

```python
raw_data.isnull().sum()
```

```python
# Insert 'Currency' column with value 'МКД' after 'Discount Price'
discount_index = raw_data.columns.get_loc('Discount Price')
raw_data.insert(discount_index + 1, 'Currency', 'МКД')
```

```python
raw_data.head(5)
```

```python
# Convert to list
raw_data['Available Sizes'] = raw_data['Available Sizes'].apply(lambda x: [s.strip() for s in x.split(',') if s.strip()])
```

```python
raw_data.head(5)
```

```python
#Consistent text formatting improves grouping, filtering, and avoids mismatches.
#Convert text to lowercase.
#Strip extra whitespace.
#Optionally remove special characters.

text_columns = ['Name', 'Brand', 'Description','Currency']

for col in text_columns:
    raw_data[col] = raw_data[col].str.strip().str.lower()
```

```python
raw_data.head(5)
```

```python
raw_data = raw_data.rename(columns={'Image URL': 'Линк до сликата на продуктот', 'Name': 'Име', 'Regular Price':'Регуларна цена', 'Discount Price':'Намалена цена', 'Currency':'Валута', 'Available Sizes':'Достапни величини','Brand':'Бренд', 'Code':'Код на продуктот', 'Description':'Опис'})
```

```python
raw_data.head(5)
```

```python
name_encoding = {
    'јакна': 1,
    'кошула': 2,
    'бермуди': 3,
    'одело (палто)': 4,
    'поло маица': 5,
    'фармерки': 6
}

brand_encoding = {
    'timberland': 1,
    'dan john': 2,
    'guess': 3,
    'boss': 4,
    'calvin klein jeans': 5,
    'replay': 6,
    "levi's": 7,
    'mango': 8,
    'tommy jeans': 9,
    'hugo': 10
    # Add more brands if present in your data
}

# Apply the encoding to the "Име" column and convert to integer
raw_data['Име'] = raw_data['Име'].map(name_encoding).fillna(raw_data['Име']).astype(int)

# Apply the encoding to the "Бренд" column and convert to integer
raw_data['Бренд'] = raw_data['Бренд'].map(brand_encoding).fillna(raw_data['Бренд']).astype(int)
```

```python
raw_data.head(5)
```

```python
for col in ['Регуларна цена','Намалена цена']:
    raw_data[col] = (
        raw_data[col]                    # original strings
          .astype(str)                   # make sure it’s text
          .str.replace(r'[^0-9]', '', regex=True)  # drop anything that isn’t 0–9
          .astype(int)                   # convert “1290” → 1290
    )
```

```python
raw_data
```

```python
raw_data.to_csv('fashiongroup_maski_cleaned.csv', index=False)
```

```python
import pandas as pd
import ast, json
```

```python
stand_data = pd.read_csv('fashiongroup_maski_cleaned.csv')
stand_data
```

```python
stand_data['sizes_list'] = stand_data['Достапни величини'].apply(ast.literal_eval)
```

```python
stand_data['std_sizes'] = stand_data['sizes_list'].apply(lambda L: [s.strip().upper() for s in L])
```

```python
def make_attributes(row):
    return {
        "име":                row['Име'],
        "регуларна_цена":     row['Регуларна цена'],
        "намалена_цена":      row['Намалена цена'],
        "валута":             row['Валута'],
        "достапни_величини":  row['std_sizes'],
        "бренд":              row['Бренд'],
        "код":                row['Код на продуктот'],
        "опис":               row['Опис'],
        "линк_слика":         row['Линк до сликата на продуктот']
    }
```

```python
stand_data['attributes']      = stand_data.apply(make_attributes, axis=1)
stand_data['attributes_json'] = stand_data['attributes'].apply(lambda d: json.dumps(d, ensure_ascii=False))
```

```python
print(stand_data[['Име','attributes_json']].head())
```

```python
import matplotlib.pyplot as plt

cleaned_fsh_maski['Величини_list'] = cleaned_fsh_maski['Достапни величини'].apply(ast.literal_eval)

# Додавање колона со број на достапни величини
cleaned_fsh_maski['Број_величини'] = cleaned_fsh_maski['Величини_list'].apply(len)

cleaned_fsh_maski['Попуст'] = cleaned_fsh_maski['Регуларна цена'] - cleaned_fsh_maski['Намалена цена']
```

```python
def show():
    plt.tight_layout()
    plt.show()
```

```python
# 1. Хистограм на регуларните цени
plt.figure(figsize=(8,5))
plt.hist(cleaned_fsh_maski['Регуларна цена'].dropna(), bins=30)
plt.title('Распределба на регуларните цени')
plt.xlabel('Регуларна цена')
plt.ylabel('Број на производи')
show()
```

```python
# 2. Хистограм на намалените цени
plt.figure(figsize=(8,5))
plt.hist(cleaned_fsh_maski['Намалена цена'].dropna(), bins=30)
plt.title('Распределба на намалените цени')
plt.xlabel('Намалена цена')
plt.ylabel('Број на производи')
show()
```

```python
# 3. Хистограм на апсолутниот попуст
plt.figure(figsize=(8,5))
plt.hist(cleaned_fsh_maski['Попуст'].dropna(), bins=30)
plt.title('Распределба на апсолутниот попуст')
plt.xlabel('Попуст (Регуларна - Намалена)')
plt.ylabel('Број на производи')
show()
```

```python
# 4. Бокс-плот на регуларните цени
plt.figure(figsize=(6,5))
plt.boxplot(cleaned_fsh_maski['Регуларна цена'].dropna(), vert=True)
plt.title('Варијација на регуларните цени')
plt.ylabel('Регуларна цена')
show()
```

```python
# 5. Бокс-плот на намалените цени
plt.figure(figsize=(6,5))
plt.boxplot(cleaned_fsh_maski['Намалена цена'].dropna(), vert=True)
plt.title('Варијација на намалените цени')
plt.ylabel('Намалена цена')
show()
```

```python
# 6. Скатер: Регуларна vs. Намалена цена
plt.figure(figsize=(8,5))
plt.scatter(cleaned_fsh_maski['Регуларна цена'], cleaned_fsh_maski['Намалена цена'])
plt.title('Регуларна цена vs. Намалена цена')
plt.xlabel('Регуларна цена')
plt.ylabel('Намалена цена')
show()
```

```python
# 7. Скатер: Намалена цена vs. број достапни величини
plt.figure(figsize=(8,5))
plt.scatter(cleaned_fsh_maski['Број_величини'], cleaned_fsh_maski['Намалена цена'])
plt.title('Намалена цена vs. Број на достапни величини')
plt.xlabel('Број на величини')
plt.ylabel('Намалена цена')
show()
```

```python
# 8. Бар-график: фреквенција на достапни величини
all_sizes = [size for sublist in cleaned_fsh_maski['Величини_list'] for size in sublist]
size_counts = pd.Series(all_sizes).value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.bar(size_counts.index, size_counts.values)
plt.title('Фреквенција на достапни величини')
plt.xlabel('Величина')
plt.ylabel('Број појавувања')
plt.xticks(rotation=45)
show()
```

```python
# 9. Бар-график: број на производи по бренд
brand_counts = cleaned_fsh_maski['Бренд'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(brand_counts.index.astype(str), brand_counts.values)
plt.title('Број на производи по бренд')
plt.xlabel('Бренд')
plt.ylabel('Број на производи')
plt.xticks(rotation=45)
show()
```
